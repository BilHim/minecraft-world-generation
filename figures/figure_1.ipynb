```python
from PIL import Image, ImageDraw
from matplotlib import pyplot as plt
import numpy as np
from scipy.spatial import cKDTree
```

```python
width, height = 3200, 1600

points = np.array([
    [1600, 200],
    [2400, 1120],
    [800, 1320]
])

colors = [
    (12, 236, 221),
    (255, 243, 56),
    (196, 0, 255)
]

tree = cKDTree(points)
```

```python
for t in range(100):
    im = Image.new('RGB', (width, height), (255, 255, 255))
    draw = ImageDraw.Draw(im)

    theta = t/100*2*np.pi
    print(t, theta/2/np.pi)
    p = np.array([1600 , 1000]) + 700*np.array([np.cos(theta), np.sin(2*theta)/2]) 
    d_min, i_min = tree.query(p)

    for i, point in enumerate(points):
        p1 = point
        c = (240, 240, 240) if i != i_min else (100, 100 ,100)
        draw.line((*p1, *p), fill=c, width=10)

    circle_size = np.array([50, 50])
    p1 = p - circle_size/2
    p2 = p + circle_size/2  
    draw.ellipse((*p1, *p2), fill=colors[i_min])

    for i, point in enumerate(points):
        circle_size = np.array([100, 100])
        p1 = point - circle_size/2
        p2 = point + circle_size/2
        
        draw.ellipse((*p1, *p2), fill=colors[i])

    im = im.resize((800, 400), resample=Image.ANTIALIAS)
    im.save(f"out_{t}.png")
```
